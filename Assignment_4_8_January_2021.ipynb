```python
import tkinter
from tkinter import *
import sys
import os

if os.environ.get('DISPLAY','') == '':
    print('no display found. Using :0.0')
    os.environ.__setitem__('DISPLAY', ':0.0')
```

```python
# Create main window
master = Tk()
master.title("Jayesh's Cryptography App")
master.geometry("300x100")

def encryptMessage():
    a = var.get()
    ct = onetimepad.encrypt(a,"cryptography")
    print("Encrypted",ct) 
    
    e2.delete(0,END)
    e2.insert(END,ct)

def dycrptMessage():
    a = var2.get()
    ct = onetimepad.decrypt(a,"cryptography")
    print("Decrypted",ct) 
    
    e4.delete(0,END)
    e4.insert(END,ct)
    
l1 = Label(root,text="Message to encrypt")
l1.grid(row=0,column=0)

l2 = Label(root,text="Encypted message")
l2.grid(row=1,column=0)

l3 = Label(root,text="Encypted message")
l3.grid(row=0,column=2)

l4 = Label(root,text="Decrypted message")
l4.grid(row=1,column=2)

var= StringVar()
e1 = Entry(master,textvariable=var)
e1.grid(row=0,column=1)

var2= StringVar()
e3 = Entry(master,textvariable=var2)
e3.grid(row=0,column=3)

e2 = Entry(master)
e2.grid(row=1,column=1)

e4 = Entry(master)
e4.grid(row=1,column=3)

b1 = Button(root,text="Encypt",bg="grey",fg="Red",command=encryptMessage)
b1.grid(row=2,column=1)

b2 = Button(root,text="Decrpt",bg="blue",fg="Red",command=dycrptMessage)
b2.grid(row=2,column=3)

root.mainloop()
```

### Guess the name of the movie

```python
import random
movie = database["Name"]
player = input("What's your Name: ")
print("Guess a movie character: ")
print("You have 10 chance to get the correct movie name: ")
print("Best of luck!",player)


count = 10
geuss= ""
word = random.choice(movie) #war

while count>0:
    fail = 0
    for char in word:
        if char in geuss:
            print(char)
        else:
            print("_")
            fail+=1
                        
    if fail==0:
        print("Congratulation you won!!!")
        print("Correct movie name was: ",word)
        break
        
    g = input("Enter a character of movie name: ")
    geuss = geuss+g
    
    if g not in word:
        count = count-1
        print("Incorrect answer:( ")

        print("You have ",count,"more geusses. Think carefully!")
```

## Project 4
Sending mail using Python

```python
import smtplib, ssl

port = 465  # For SSL
gmail_user = 'from_jayesh@gmail.com'
gmail_password = input("Type your password and press enter: ")

# Create a secure SSL context
context = ssl.create_default_context()

with smtplib.SMTP_SSL("smtp.gmail.com", port, context=context) as server:
    server.login("jayesh.dsaiml@gmail.com", password)
    # TODO: Send email here


sent_from = gmail_user
to = ['to_jayesh@gmail.com']
subject = 'OMG Super Important Message'
body = "Hey, what's up?\n\n- You"

email_text = """\
From: %s
To: %s
Subject: %s

%s
""" % (sent_from, ", ".join(to), subject, body)

try:
    server = smtplib.SMTP_SSL('smtp.gmail.com', 465)
    server.ehlo()
    server.login(gmail_user, gmail_password)
    server.sendmail(sent_from, to, email_text)
    server.close()

    print('Email sent!')
except:
    print('Something went wrong...')
```
